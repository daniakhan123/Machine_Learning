```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("cancer.csv")
print("Unique Target Categories:", df['Level'].unique())
print("Total columns:", len(df.columns))
print(df.head(50))
print(df.tail(30))
df.info()
print("Shape:", df.shape)
print("Size:", df.size)
print("Number of samples (rows):", df.shape[0])
print("Number of columns:", df.shape[1])
print(df['Level'].value_counts())
```

```python
sns.countplot(x='Level', data=df)
plt.title("Distribution of Target Variable")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
sample1 = df.loc[550:900, 'Gender':'Coughing of Blood']
sample1.to_csv("extracted_sample1.csv", index=False)


print("Mean:\n", sample1.mean())
print("Median:\n", sample1.median())
print("Mode:\n", sample1.mode().iloc[0])
print("Std Dev:\n", sample1.std())
print("Min:\n", sample1.min())
print("Max:\n", sample1.max())
```

```python

sample2 = df.iloc[-20:, df.columns.get_loc('Shortness of Breath'):df.columns.get_loc('Frequent Cold')+1]
sample2.to_csv("extracted_sample2.csv", index=False)


print("Mean:\n", sample2.mean())
print("Median:\n", sample2.median())
print("Mode:\n", sample2.mode().iloc[0])
print("Std Dev:\n", sample2.std())
print("Min:\n", sample2.min())
print("Max:\n", sample2.max())
```
