```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.utils import resample
from sklearn.feature_selection import VarianceThreshold


data = pd.read_csv("train.csv")
print("Dataset Shape:", data.shape)
print("Columns:", data.columns.tolist())


cols = ['var3', 'var15', 'var38', 'imp_op_var39_comer_ult1']
cols = [c for c in cols if c in data.columns]
corr = data[cols].corr()
print("\nCorrelation among selected variables:\n", corr)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

sns.pairplot(data[cols].sample(2000))  
plt.show()


print("\nMissing values before imputation:\n", data.isnull().sum().head())
data.fillna(data.mean(numeric_only=True), inplace=True)
print("\nMissing values after imputation:\n", data.isnull().sum().head())


sns.histplot(data['var38'], kde=True)
plt.title("Distribution of var38")
plt.show()

sns.boxplot(x='var15', y='var38', data=data)
plt.title("Boxplot of var38 vs var15")
plt.show()

sns.scatterplot(x='var3', y='var38', data=data.sample(2000))
plt.title("Relationship between var3 and var38")
plt.show()

sns.countplot(x='var15', data=data)
plt.title("Frequency of var15 Categories")
plt.show()

sns.heatmap(data[cols].corr(), annot=True, cmap='coolwarm')
plt.title("Heatmap of Selected Variables")
plt.show()

target = 'TARGET'
print("\nTarget variable distribution:")
print(data[target].value_counts())
sns.countplot(x=target, data=data)
plt.title("Target Variable Distribution")
plt.show()


majority = data[data[target] == 0]
minority = data[data[target] == 1]
minority_upsampled = resample(minority, replace=True, n_samples=len(majority), random_state=42)
balanced = pd.concat([majority, minority_upsampled])
print("\nAfter upsampling:\n", balanced[target].value_counts())


corr_matrix = data.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.65)]
pearson_selected = data.drop(columns=to_drop)
print("\nAfter Pearson Feature Selection:", pearson_selected.shape)

selector = VarianceThreshold(threshold=0.01)
selected = selector.fit_transform(data.select_dtypes(include=np.number))
print("After Variance Threshold Feature Selection:", selected.shape)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

data = pd.DataFrame({
    'Age':[18,22,25,30,35,40,28,26,33,45],
    'Gender':['M','F','M','F','M','F','F','M','M','F'],
    'Satisfaction':[3,4,5,2,1,5,4,3,2,1],
    'Recommend':['Yes','Yes','No','No','Yes','Yes','No','Yes','No','No']
})

data['Gender'] = LabelEncoder().fit_transform(data['Gender'])
data['Recommend'] = LabelEncoder().fit_transform(data['Recommend'])
X = data[['Age','Gender','Satisfaction']]
y = data['Recommend']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

kernels = ['linear','poly','rbf','sigmoid']
for k in kernels:
    svm = SVC(kernel=k)
    svm.fit(X_train, y_train)
    print(k, accuracy_score(y_test, svm.predict(X_test)))

svm_default = SVC()
svm_tuned = SVC(C=2, gamma=0.5, kernel='rbf')
svm_default.fit(X_train, y_train)
svm_tuned.fit(X_train, y_train)
print('Default Params:', accuracy_score(y_test, svm_default.predict(X_test)))
print('Tuned Params:', accuracy_score(y_test, svm_tuned.predict(X_test)))
```
